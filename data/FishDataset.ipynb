```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fishData = pd.read_csv("fish.csv")
fishData
```

```python
plt.scatter(fishData["Weight"],fishData["Length1"], label=fishData["Species"])
plt.show()
```

```python
import seaborn as sns
ax = sns.scatterplot(x="Weight", y="Length1",
                     hue="Species",
#                      style="Species", 
                     data=fishData)
ax.set_ylabel("Vertical Length (cm)")
ax.set_xlabel("Weight (g)")
h, l = ax.get_legend_handles_labels()
h, l = h[1:], l[1:]
ax.legend(h, l, loc='lower left', bbox_to_anchor= (0.01, 1.01), ncol=4,
            borderaxespad=0, frameon=True, fontsize=11)
plt.savefig("weight-vs-vertical_length.png")
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
np.random.seed(1974)

font = {'size'  : 12}

matplotlib.rc('font', **font)

# Generate Data
x = fishData["Weight"]
y = fishData["Length1"]
labels = fishData["Species"]
df = pd.DataFrame(dict(x=x, y=y, label=labels))

groups = df.groupby('label')

# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.x, group.y, marker='+', linestyle='', ms=12, label=name)  
plt.legend(title="Species")
plt.xlabel("Weight (g)")
plt.ylabel("Length1: Vertical Length (cm)")
plt.savefig('Length1.png')
plt.show()
```
