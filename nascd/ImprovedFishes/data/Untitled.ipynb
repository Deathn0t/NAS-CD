```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_ppo2 = pd.read_csv("training_best_ppo2.csv")
df_ppo3 = pd.read_csv("training_best_ppo2_10.csv")
df_base = pd.read_csv("training_baseline.csv")
```

```python
plt.plot(df_ppo2.val_loss, label="ppo2")
plt.plot(df_base.val_loss, label="base")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
plt.plot(df_ppo2.val_r2, label="ppo2")
plt.plot(df_ppo3.val_r2, label="ppo2+2")
plt.plot(df_base.val_r2, label="base")
plt.legend()
plt.show()
```
